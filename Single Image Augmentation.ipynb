```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc, ndimage
import imageio
import keras
%matplotlib inline
```

```python
gen=ImageDataGenerator(
    rotation_range=10,width_shift_range=0.1,
    height_shift_range=0.1,shear_range=0.15,
    zoom_range=0.1,
    channel_shift_range=10,
    horizontal_flip=True
)
```

```python
image_path=r'C:\Users\STACBD\bangla-money\Training\1\1_0.jpg'
```

```python
image=np.expand_dims(imageio.imread(image_path),0)
plt.imshow(image[0])
```

```python
aug_iter=gen.flow(image,save_to_dir=r'C:\Users\STACBD\bangla-money\Save', save_prefix='1_', save_format='jpg')
```

```python
aug_images=[next(aug_iter)[0].astype(np.uint8) for i in range(15)]
```
